```python
import yfinance as yf
import pandas as pd
import os

def get_stock_data(tickers, start_date, end_date):
    if isinstance(tickers, str):
        tickers = [tickers]
    stock_data = yf.download(tickers, start=start_date, end=end_date)
    return stock_data

tickers = ['VOO', 'SPY', 'IVV', 'SWPPX', 'QQQ', 'VTWO', 'VTSAX', 'DIA']
start_date = '2019-01-01'
end_date = '2024-02-21'

# Fetch the stock data
stock_data = get_stock_data(tickers, start_date, end_date)

# Flatten the column MultiIndex for CSV
if isinstance(stock_data.columns, pd.MultiIndex):
    stock_data.columns = [' '.join(col).strip() for col in stock_data.columns.values]

# Define the full path for the CSV file
csv_file_path = r'C:\Users\mikea\Desktop\Dataclass2\project-2-stock-data\Resources\stock_data.csv'

# Check if the directory exists
if os.path.exists(os.path.dirname(csv_file_path)):
    print("Directory exists. Saving file...")
    stock_data.to_csv(csv_file_path)
    print(f"Stock data has been successfully saved to {csv_file_path}")
else:
    print(f"Directory does not exist: {os.path.dirname(csv_file_path)}")
```

```python
import pandas as pd

csv_file_path = r'C:\Users\mikea\Desktop\Dataclass2\project-2-stock-data\Resources\stock_data.csv'

stock_data_df = pd.read_csv(csv_file_path)
stock_data_df.set_index('Date', inplace=True)

# Round columns 'Adj Close DIA', 'Adj Close IVV' and 46 other columns (Number of decimals: 2)
stock_data_df = stock_data_df.round({'Adj Close DIA': 2, 'Adj Close IVV': 2, 'Adj Close QQQ': 2, 'Adj Close SPY': 2, 'Adj Close SWPPX': 2,
                                     'Adj Close VOO': 2, 'Adj Close VTSAX': 2, 'Adj Close VTWO': 2, 'Close DIA': 2, 'Close IVV': 2, 'Close QQQ': 2,
                                     'Close SPY': 2, 'Close SWPPX': 2, 'Close VOO': 2, 'Close VTSAX': 2, 'Close VTWO': 2, 'High DIA': 2, 'High IVV': 2,
                                     'High QQQ': 2, 'High SPY': 2, 'High SWPPX': 2, 'High VOO': 2, 'High VTSAX': 2, 'High VTWO': 2, 'Low DIA': 2,
                                     'Low IVV': 2, 'Low QQQ': 2, 'Low SPY': 2, 'Low SWPPX': 2, 'Low VOO': 2, 'Low VTSAX': 2, 'Low VTWO': 2, 'Open DIA': 2,
                                     'Open IVV': 2, 'Open QQQ': 2, 'Open SPY': 2, 'Open SWPPX': 2, 'Open VOO': 2, 'Open VTSAX': 2, 'Open VTWO': 2,
                                     'Volume DIA': 2, 'Volume IVV': 2, 'Volume QQQ': 2, 'Volume SPY': 2, 'Volume SWPPX': 2, 'Volume VOO': 2,
                                     'Volume VTSAX': 2, 'Volume VTWO': 2})

stock_data_df.head()
```

```python
import matplotlib.pyplot as plt

# extract ticker code from the column names:
ticker_code= [name.split()[-1] for name in stock_data_df.columns]
#new DF with the ticker code as column names
df_ticker= pd.DataFrame(stock_data_df.values, columns= ticker_code, index= stock_data_df.index)


# create graphs.
for ticker_code in df_ticker.columns:
    plt.figure(figsize=(10, 4))
    plt.plot(df_ticker.index, df_ticker[ticker_code], label=ticker_code) 
    plt.title(f' Stock price of {ticker_code} Over Time')
    plt.xlabel('Month')
    plt.ylabel('Adjusted Close Price')
    plt.legend()
    plt.show()
    
```
